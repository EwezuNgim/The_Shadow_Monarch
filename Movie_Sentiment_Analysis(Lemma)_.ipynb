**Data** **Cleaning**

```python
def clean_punctuations(data):
 for i in range(0,len(data)):
   word=data['en_review'][i]
   for item in punctuations:
     word= word.replace(item,'')
   data['en_review'][i]=word
   #punctuations removed
   #success
   corpus=[]
 #print("Here it comes, no punctuations:")
 #print(data['en_review'])
 for i in range(0,len(data)):
       #corpus.extend(train['en_review'][i].split(' '))
       corpus.append(data['en_review'][i])
       #punctuationless text stored
 nltk.download('stopwords')
 corpus_v2=[item for item in corpus if item.lower() not in set(stopwords.words('english'))]
 #store all reviews after stopwords have been removed
 nltk.download('wordnet')
 lemma=WordNetLemmatizer()
 clean_data=np.array([lemma.lemmatize(word) for word in corpus_v2])
 return clean_data
```
