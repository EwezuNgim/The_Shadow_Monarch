```python
from duckduckgo_search import ddg_images
from fastcore.all import*
```

```python
def search_images(item,max_images=5):
    print(f"Searching for {item}...")
    return L(ddg_images(item,max_results=max_images)).itemgot('image')
```

```python
link=search_images('polar bear')
```

```python
from fastdownload import download_url
dest='P_Bear.jpg'
download_url(link[0],dest,show_progress=False)
```

```python
from fastai.vision.all import *
```

```python
pic=Image.open(dest)
pic.to_thumb(256)
```

```python
search='polar bear','grizzly bear','teddy bear'
path=Path('bears')
from time import sleep
for i in search:
    dest=(path/i)
    dest.mkdir(exist_ok=True,parents=True)
    download_images(dest,urls=search_images(f"{i} indoors"))
    sleep(10)
    download_images(dest,urls=search_images(f"{i} outdoors"))
    sleep(10)
    resize_images(path/i,max_size=300,dest=path/i)
```

```python
failed_img=verify_images(get_image_files(path))
#verify checks a list consisting of the images in that location 
#returns True or false for each image
failed_img.map(Path.unlink)
#removes invalid images
len(failed_img)
```

```python
dls=DataBlock(blocks=(ImageBlock,CategoryBlock),get_items=get_image_files,splitter=RandomSplitter(valid_pct=0.2,seed=42),get_y=parent_label,item_tfms=Resize(192)).dataloaders(path,bs=32)
#dataloaders passes the training and val set in batches to the GPU
```

```python
model=vision_learner(dls,resnet18,metric=error_rate)
```

```python
model.fine_tune(2)
```
